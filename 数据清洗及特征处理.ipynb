```python
import numpy as np
import pandas as pd
```

```python
#加载数据train.csv
df = pd.read_csv('D:/桌面/资料/DBA/train.csv')
df.head(3)
```

```python
#缺失值观察
df.isnull().sum()
```

```python
#对缺失值进行处理
df[df['Age'].isnull()] = 0
df.head()
```

```python
df.dropna().head(3)
```

```python
df.fillna(0).head(3)
```

```python
#查看重复值
df[df.duplicated()]
```

```python
#处理重复值
df = df.drop_duplicates()
#删除 DataFrame 中的重复行
df.head()
```

```python
df.to_csv('D:/桌面/资料/DBA/train_clear.csv')
```

```python
#对年龄进行分箱操作
#将连续变量Age平均分箱成5个年龄段，并分别用类别变量12345表示
df['AgeBand'] = pd.cut(df['Age'],5,labels = [1,2,3,4,5])
df.head()
```

```python
#保存
df.to_csv('D:/桌面/资料/DBA/train_ave.csv')
```

```python
#将连续变量Age划分为(0,5] (5,15] (15,30] (30,50] (50,80]五个年龄段，并分别用类别变量12345表示
df['AgeBand'] = pd.cut(df['Age'],[0,5,15,30,50,80],labels = [1,2,3,4,5])
df.head(3)
```

```python
#保存
df.to_csv('D:/桌面/资料/DBA/train_cut.csv')
```

```python
#将连续变量Age按10% 30% 50 70% 90%五个年龄段，并用分类变量12345表示
df['AgeBand'] = pd.qcut(df['Age'],[0,0.1,0.3,0.5,0.7,0.9],labels = [1,2,3,4,5])
df.head(3)
```

```python
#保存
df.to_csv('D:/桌面/资料/DBA/train_pr.csv')
```

```python
#将类别文本转换为12345

#方法一: replace
df['Sex_num'] = df['Sex'].replace(['male','female'],[1,2])
df.head()
```

```python
#将类别文本转换为one-hot编码

#方法一: OneHotEncoder
for feat in ["Age", "Embarked"]:
#     x = pd.get_dummies(df["Age"] // 6)
#     x = pd.get_dummies(pd.cut(df['Age'],5))
    x = pd.get_dummies(df[feat], prefix=feat)
    df = pd.concat([df, x], axis=1)
    #df[feat] = pd.get_dummies(df[feat], prefix=feat)

df.head()
```

```python
df['Title'] = df.Name.str.extract('([A-Za-z]+)\.', expand=False)
df.head()
```
